```python
# 3.10 Window / Ranking in Pandas

# 1. For each city, rank transactions by amount descending


merged_df["amount_rank"] = (merged_df.groupby("city")["amount"].rank(method="dense",ascending= False))
merged_df
```

```python
# 2. Get top 2 transactions per city (include ties using dense ranking)

merged_df["amount_rank"] = ( merged_df.groupby("city")["amount"].rank(method="dense",ascending=False))

top_2_per_city = merged_df[merged_df["amount_rank"] <=2]
top_2_per_city
```

```python
# 3.11 Rolling / Expanding

# 1. Rolling mean of amount per customer (window=3)


#Step 1: Sorting

# merged_df = merged_df.sort_values(["customer_id","txn_date"])

# #step 2: Rolling mean ( window = 3)

# merged_df["rolling_mean_3"] = ( merged_df.groupby("customer_id")["amount"].rolling(window=3).mean().reset_index(level=0,drop=True))
# merged_df

merged_df["rolling_mean_3"] = (
    merged_df
    .groupby("customer_id")["amount"]
    .rolling(window=3, min_periods=1)
    .mean()
    .reset_index(level=0, drop=True)
)
merged_df
```

```python
# 2. Cumulative sum of amount per customer

#Step 1: sort properly
merged_df = merged_df.sort_values(["customer_id","txn_date"])

#Step 2: Cumulative sum per customer

merged_df["cumulative_amount"] = ( merged_df.groupby("customer_id")["amount"].cumsum())
merged_df
```
